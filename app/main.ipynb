```python
import subprocess
```

```python
# Cas 1 : commande qui réussit
result = subprocess.run("echo hello", shell=True, capture_output=True, text=True)
print("stdout:", repr(result.stdout))
print("stderr:", repr(result.stderr))
```

```python
# Cas 2 : commande qui échoue
result = subprocess.run("cat fichier_inexistant.txt", shell=True, capture_output=True, text=True)
print("stdout:", repr(result.stdout))
print("stderr:", repr(result.stderr))
```
